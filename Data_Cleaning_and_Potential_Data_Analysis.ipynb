```python
def counts_stuff(col):
    count_of_string_values =  0
    count_of_folat_values = 0 
    for i in col:
        if isinstance(i,str):
            count_of_string_values +=1
        if isinstance(i,float): 
            count_of_folat_values +=1


    print ("Count of values that are considered floats: ",count_of_folat_values)
    print ("count of string values that are considered strings:", count_of_string_values)
    # as we can see all of the na values are considered floats, hence we need to drop them

counts_stuff(newdf['Records'])
```

```python
cleanednewdf = newdf.dropna()
```

```python
counts_stuff(cleanednewdf['Records'])
```

```python
cleanednewdf = cleanednewdf.replace(',','', regex=True)
```

```python
def convert_to_int_if_number(s):
    if bool(re.match(r'^\d{1,3}(,\d{3})*(\.\d+)?$|^\d+(\.\d+)?$', s)):
        # Remove commas and convert to int
        return int(s.replace(',', ''))
    return pd.NA
```

```python
cleanednewdf['Records'] = cleanednewdf['Records'].apply(convert_to_int_if_number)
```

```python
count_of_none_numbers = cleanednewdf.isna().sum()
count_of_none_numbers
```
